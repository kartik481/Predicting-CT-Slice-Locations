```python
import mysql.connector
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk import tokenize
import sys
import nltk
nltk.download('vader_lexicon')
nltk.download('punkt')
```

```python
db= mysql.connector.connect(user='root', password='password',
                              host='127.0.0.1',database="online4"                             
                              )
cursor = db.cursor(buffered=True)
```

```python
updateScoreQuery = "UPDATE sentiment \
SET sentiment_score = %s, \
max_pos_score = %s, \
max_neg_score = %s \
WHERE url = %s"
```

```python
selectEmailQuery = "SELECT url, body FROM sentiment"
cursor.execute(selectEmailQuery)
emails = cursor.fetchall()
```

```python
for email in emails:
    url = email[0]
    body = email[1]
    finalScore = 0
    roundedFinalScore = 0
    maxPosScore = 0
    maxNegScore = 0
    print("===")
    sid = SentimentIntensityAnalyzer()
    emailLines = tokenize.sent_tokenize(body)
    for line in emailLines:
        ss = sid.polarity_scores(line)
        line = line.replace('\n', ' ').replace('\r', '')
        print(line)
        for k in sorted(ss):
            print(' {0}: {1}\n'.format(k,ss[k]), end='')
        lineCompoundScore = ss['compound']
        finalScore += lineCompoundScore
        if ss['pos'] > maxPosScore:
            maxPosScore = ss['pos']
        elif ss['neg'] > maxNegScore:
            maxNegScore = ss['neg']
            
    # calculate avg compound score for the entire message
    roundedFinalScore = round(finalScore/len(emailLines),4)
    print("***Final Email Score", roundedFinalScore)
    print("Most Positive Sentence Score:", maxPosScore)
    print("Most Negative Sentence Score:", maxNegScore)
    # update table with calculated fields
    try:
        cursor.execute(updateScoreQuery,(roundedFinalScore,maxPosScore, maxNegScore, url))
        db.commit()
    except:
        db.rollback()
        db.close()
```
